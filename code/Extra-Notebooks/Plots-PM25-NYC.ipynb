```python
# Combine plots for each type
def plot_combined_data(resolutions, rmse_data, mape_data, color, title_prefix, resolution_phrase='Resolution'):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
    ax1.plot(resolutions, rmse_data, 'o-', color=color)
    ax1.set_title(f'{title_prefix} RMSE vs. {resolution_phrase}')
    ax1.set_xlabel(resolution_phrase)
    ax1.set_ylabel('RMSE')
    ax1.grid(True)
    
    ax2.plot(resolutions, mape_data, 'o-', color=color)
    ax2.set_title(f'{title_prefix} MAPE vs. {resolution_phrase}')
    ax2.set_xlabel(resolution_phrase)
    ax2.set_ylabel('MAPE (%)')
    ax2.grid(True)
    
    plt.tight_layout()
    plt.show()

# Plotting
plot_combined_data(resolutions_geohash, rmse_geohash, mape_geohash, 'blue', 'Geohash', 'Precision')
plot_combined_data(resolutions_h3, rmse_h3, mape_h3, 'green', 'H3')
plot_combined_data(resolutions_s2, rmse_s2, mape_s2, 'red', 'S2')
```
